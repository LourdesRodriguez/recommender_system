```python
df_bgame = pd.read_csv(folder+'/BggGames.csv', usecols= ['bgg_id','name','category','rank'])

df_bgame.head(10)
```

```python
df_bgame100= df_bgame[df_bgame['rank']<=100].sort_values('rank')
df_bgame100
```

```python
category_num= ','.join(df_bgame100.category).split(',')

print(np.unique(category_num))
print('Total Categories in Top 100 Boardgames: ' + str(len(np.unique(category_num))))
```

```python

df_cat = pd.read_csv(folder+'/CategoriesID.csv',sep=';', usecols=['Category','IdCat'])
df_cat.head(10)
```

```python
category = list(df_cat.Category)
idcategory= list(df_cat.IdCat)
dic_category = dict(zip(idcategory,category))
dic_category
```

```python
top100cat= np.unique(category_num)
ch_top100cat= [dic_category.get(n, n) for n in top100cat]
ch_top100cat.sort()
ch_top100cat
```
