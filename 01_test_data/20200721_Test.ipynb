```python
import warnings
warnings.filterwarnings("ignore")

vector_features=df_category[['Abstract Strategy', 'Action / Dexterity',
       'Adventure', 'Age of Reason', 'American West', 'Ancient', 'Animals',
       'Arabian', 'Aviation / Flight', 'Bluffing', 'Card Game',
       'City Building', 'Civil War', 'Civilization', 'Collectible Components',
       'Comic Book / Strip', 'Deduction', 'Dice', 'Economic', 'Educational',
       'Environmental', 'Exploration', 'Fantasy', 'Farming', 'Fighting',
       'Horror', 'Industry / Manufacturing', 'Mature / Adult', 'Medical',
       'Medieval', 'Miniatures', 'Modern Warfare', 'Movies / TV / Radio theme',
       'Murder/Mystery', 'Mythology', 'Negotiation', 'Novel-based',
       'Party Game', 'Pike and Shot', 'Political', 'Post-Napoleonic',
       'Prehistoric', 'Puzzle', 'Religious', 'Renaissance', 'Science Fiction',
       'Space Exploration', 'Spies/Secret Agents', 'Territory Building',
       'Trains', 'Transportation', 'Travel', 'Video Game Theme', 'Wargame',
       'Word Game']]

vector_features['features'] = vector_features.apply(lambda x: np.array(x), axis=1)
df_category_vf = df_category[['bgg_id', 'name']]
df_category_vf['features'] = vector_features['features']
df_category_vf
```

```python
i = 77695 
game = df_category_vf.name[i]
print('Similar Boardgames: ' + str(game))

x = df_category_vf.features[i]

df_category_vf['euclidean'] = df_category_vf.features.apply(lambda y: np.sqrt(np.dot(x, x) - 2 * np.dot(x, y) + np.dot(y, y)))
df_category_vf['cosine'] = df_category_vf.features.apply(lambda y: np.dot(x, y) / (np.sqrt(np.dot(x,x)) * np.sqrt(np.dot(y,y))))

df_category_vf.head()
```
