```python
import pandas as pd
from rake_nltk import Rake
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
folder= './Data'
```

```python
df_bgame = pd.read_csv(folder+'/BggGames.csv', usecols= ['bgg_id','name','category','rank'])

df_bgame.head(10)
```

```python
df_bgame100= df_bgame[df_bgame['rank']<=100].sort_values('rank')
df_bgame100.tail(50)
```

```python
df_bgame100.shape
```

```python
df_bgame100['category'] = df_bgame100['category'].map(lambda x: x.replace(',', ' '))
print(df_bgame100['category'])
```

```python
df_bgame100.drop(columns = ['rank'], inplace = True)
```

```python
df_bgame100.set_index('name', inplace = True)
df_bgame100.head()
```

```python
print(np.where(df_bgame100.index=='Tal der Könige')[0])
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df_bgame100['category'])
```

```python
cosine = cosine_similarity(count_matrix, count_matrix)
cosine
```

```python
def game_recommendations(name, cosine = cosine):
    
    recommended = []
    indices = pd.Series(df_bgame100.index)
    idx = indices[indices == name].index[0]
    score = pd.Series(cosine[idx]).sort_values(ascending = False)
    top_10 = list(score.iloc[1:11].index)
    for i in top_10:
        recommended.append(list(df_bgame100.index)[i])
        
    return recommended
```

```python
game_recommendations('Agricola')
```

```python
def two_game_recommendations(name1,name2, cosine = cosine):
    
    recommended = []
    other= []
    indices = pd.Series(df_bgame100.index)
    idx1 = indices[indices == name1].index[0]
    idx2 = indices[indices == name2].index[0]
    score1 = pd.Series(cosine[idx1]).sort_values(ascending = False)
    score2 = pd.Series(cosine[idx2]).sort_values(ascending = False)
    top_100_1= list(score1.iloc[1:101].index)
    top_100_2=list(score2.iloc[1:101].index)
 

    for count,i in enumerate (top_100_1):
        for x in top_100_2:
            if x== i and count<10:
                recommended.append(list(df_bgame100.index)[i])
    return recommended
            
        
```

```python
two_game_recommendations('Gloomhaven','Agricola')
```
