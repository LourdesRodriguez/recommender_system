```python
import pandas as pd
import numpy as np
```

```python
folder= './Data'
```

```python
#Content Dataset

df_bgame = pd.read_csv(folder+'/BggGames.csv', usecols= ['bgg_id','name','category','rank'])

df_bgame.head(10)
```

```python
df_bgame100= df_bgame[df_bgame['rank']<=100].sort_values('rank')
df_bgame100.head(10)
```

```python
#Score Dataset

df_score = pd.read_csv(folder+'/20200709-000000.csv')

print('Total score: ' + str(df_score.bgg_id.nunique()))
df_score.head()
```

```python
df_score100= df_score[df_score['rank']<=100].sort_values('rank')
df_score100.head(10)
```

```python
df_game100= df_bgame100.merge(df_score100, left_on='bgg_id', right_on='bgg_id')
df_game100.head()
```

```python
category_num= ','.join(df_game100.category).split(',')
category_u = np.unique(category_num)
print(category_u)
print('Total Categories in Top 100 Boardgames: ' + str(len(np.unique(category_num))))
```

```python
df_category= df_game100.copy(deep= True)
df_category
```

```python
item_feature = df_category.copy()

for i in range(len(category_u)):
    cat = category_u[i]
    item_feature[cat] = 0
    item_feature[cat] = item_feature.apply(lambda x: x.score if (cat in x.category) else 0, axis=1)

item_feature = item_feature.drop(['name','category', 'rank_x', 'rank_y', 'score'], axis=1)
item_feature = item_feature.drop_duplicates()
print(len(item_feature))
item_feature.head(5)
```

```python
vf = item_feature.drop(['bgg_id'], axis=1).to_numpy()

item_feature['feature'] = [row for row in vf]
item_feature = item_feature[['bgg_id', 'feature']]
item_feature.head(5)
```

```python
item_feature['feature'] = item_feature.feature.apply(lambda x: x / sum(x))
item_feature.head(5)
```

```python
item_feature.columns
```

```python
df_itemfeature= item_feature.merge(df_category[['bgg_id','name']], left_on='bgg_id', right_on='bgg_id')
df_itemfeature.set_index('bgg_id',inplace=True)
df_itemfeature.head(50)
```

```python
i = 230802
game = df_itemfeature.name[i]
print('Similar Boardgames: ' + str(game))

x = df_itemfeature.feature[i]

df_itemfeature['euclidean'] = df_itemfeature.feature.apply(lambda y: np.sqrt(np.dot(x, x) - 2 * np.dot(x, y) + np.dot(y, y)))
df_itemfeature['cosine'] = df_itemfeature.feature.apply(lambda y: np.dot(x, y) / (np.sqrt(np.dot(x,x)) * np.sqrt(np.dot(y,y))))

df_itemfeature.head()
```

```python
df_itemfeature[df_itemfeature['cosine']<= 1.1].sort_values('cosine', ascending=False).head(10)
```
